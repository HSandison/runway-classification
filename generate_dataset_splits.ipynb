```python
from google.colab import drive
import os
```

```python
# Mount Google Drive
drive.mount('/content/drive/')

# Set the path to the ImageSets/Main folder
main_folder_path = '/content/drive/MyDrive/FODPascalVOCFormat-V.2.1/VOC2007/ImageSets/Main/'
```

```python
path = '/content/drive/MyDrive/FODPascalVOCFormat-V.2.1/VOC2007/JPEGImages/'
os.listdir(path)
```

```python

# Get the list of image names or IDs (assuming the images are in JPEGImages folder)
jpeg_images_path = '/content/drive/MyDrive/FODPascalVOCFormat-V.2.1/VOC2007/JPEGImages/'
image_names = [name.split('.')[0] for name in os.listdir(jpeg_images_path)]

```

```python
# Set the percentage split
train_percent = 80
val_percent = 10
test_percent = 10
```

```python
# Calculate split indices
num_images = len(image_names)
train_split = int(train_percent / 100 * num_images)
val_split = int(val_percent / 100 * num_images) + train_split

```

```python
# Create and populate train.txt
train_txt_path = main_folder_path + 'train.txt'
with open(train_txt_path, 'w') as train_file:
    for name in image_names[:train_split]:
        train_file.write(name + '\n')

# Create and populate val.txt
val_txt_path = main_folder_path + 'val.txt'
with open(val_txt_path, 'w') as val_file:
    for name in image_names[train_split:val_split]:
        val_file.write(name + '\n')

# Create and populate test.txt
test_txt_path = main_folder_path + 'test.txt'
with open(test_txt_path, 'w') as test_file:
    for name in image_names[val_split:]:
        test_file.write(name + '\n')

# Print paths for debugging
print("Train file path:", train_txt_path)
print("Validation file path:", val_txt_path)
print("Test file path:", test_txt_path)
```
